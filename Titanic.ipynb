```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_cleaned = titanic.dropna()
```

```python
titanic_cleaned.isnull().sum()
```

```python
plt.figure(figsize=(12, 8))

# Age distribution
plt.subplot(2, 2, 1)
plt.hist(titanic_cleaned['Age'], bins=30, edgecolor='k')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Fare distribution
plt.subplot(2, 2, 2)
plt.hist(titanic_cleaned['Fare'], bins=30, edgecolor='k')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Sex distribution
plt.subplot(1, 2, 1)
sex_counts = titanic_cleaned['Sex'].value_counts()
plt.bar(sex_counts.index, sex_counts.values, color='skyblue', edgecolor='k')
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Frequency')

#Class distribution
plt.subplot(1,2,2)
class_counts = titanic_cleaned['Pclass'].value_counts()
plt.bar(class_counts.index.astype(str),class_counts.values,color='skyblue',edgecolor='k')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Survival rate by sex
plt.subplot(1, 2, 1)
titanic_cleaned.groupby('Sex')['Survived'].value_counts(normalize=True).unstack().plot(kind='bar', ax=plt.gca())
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Proportion')

# Survival rate by class
plt.subplot(1, 2, 2)
titanic_cleaned.groupby('Pclass')['Survived'].value_counts(normalize=True).unstack().plot(kind='bar', ax=plt.gca())
plt.title('Survival Rate by Class')
plt.xlabel('Class')
plt.ylabel('Proportion')

plt.tight_layout
```

```python
#Age v/s Fair
plt.figure(figsize=(10,6))
plt.scatter(titanic_cleaned['Age'],titanic_cleaned['Fare'])
plt.title('Age vs Fair')
plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
```

```python
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic['Embarked'].fillna('Unknown', inplace=True)  # Fill missing values with 'Unknown'
titanic['Cabin'].fillna('Unknown', inplace=True)

# Convert categorical features to numerical values for correlation calculation
titanic['Sex'] = titanic['Sex'].map({'Male': 0, 'Female': 1})
titanic['Embarked'] = titanic['Embarked'].map({'S': 0, 'C': 1, 'Q': 2, 'Unknown': -1})  # Assign a numeric value to 'Unknown'

# Select only numeric columns for correlation calculation
numeric_columns = titanic.select_dtypes(include=[np.number])

# Calculate the correlation matrix
corr_matrix = numeric_columns.corr()

# Plot the heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
